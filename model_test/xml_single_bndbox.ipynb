```python
import os
import shutil
import xml.etree.ElementTree as ET

# Source and destination directories
xml_source_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/input/label/label"  # Original directory where XML files are stored
destination_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/input/label/xml_single_bndbox"  # Destination directory for filtered XMLs

# Ensure destination directory exists
os.makedirs(destination_directory, exist_ok=True)

# Function to check if the XML contains only one bndbox
def has_single_bndbox(xml_file):
    # Parse the XML file
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Find all bounding boxes in the XML
    bndboxes = root.findall('.//bndbox')

    # Return True if there is exactly one bndbox, otherwise False
    return len(bndboxes) == 1

# Counter for XML files with a single bndbox
single_bndbox_count = 0

# Process each XML file in the source directory
for xml_filename in os.listdir(xml_source_directory):
    if xml_filename.endswith(".xml"):
        xml_file_path = os.path.join(xml_source_directory, xml_filename)

        # Check if the XML file has only one bndbox
        if has_single_bndbox(xml_file_path):
            # Move the XML file to the destination directory
            shutil.move(xml_file_path, os.path.join(destination_directory, xml_filename))
            single_bndbox_count += 1
            print(f"Moved {xml_filename} to {destination_directory}")

# Print the total number of files with a single bndbox
print(f"Total number of XML files with a single bndbox: {single_bndbox_count}")
```
