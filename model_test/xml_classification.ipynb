```python
import os
import shutil

# Source directories for XML and JPG files
xml_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/input/label/xml_single_bndbox"  # Directory where XML files are stored
jpg_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/input/images/images/water_spot"  # Directory where JPG files are stored

# Destination directory for matching XML files
xml_destination_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/input/label/water_spot"  # Directory to move matching XMLs

# Ensure destination directory exists
os.makedirs(xml_destination_directory, exist_ok=True)

# Get all JPG filenames (without extension)
jpg_filenames = {os.path.splitext(f)[0] for f in os.listdir(jpg_directory) if f.endswith(".jpg")}

# Counter for moved XML files
moved_count = 0

# Process each XML file and check for matching JPG filenames
for xml_filename in os.listdir(xml_directory):
    if xml_filename.endswith(".xml"):
        # Get the base filename (without extension)
        base_name = os.path.splitext(xml_filename)[0]

        # Check if the base name exists in the set of JPG filenames
        if base_name in jpg_filenames:
            # Move the XML file to the destination directory
            xml_file_path = os.path.join(xml_directory, xml_filename)
            shutil.move(xml_file_path, os.path.join(xml_destination_directory, xml_filename))
            moved_count += 1  # Increment the counter
            print(f"Moved {xml_filename} to {xml_destination_directory}")
        else:
            print(f"No matching JPG file found for {xml_filename}")

# Print the total number of moved XML files
print(f"Total number of moved XML files: {moved_count}")

```
