```python
import os
import zipfile
from PIL import Image, ImageDraw
import xml.etree.ElementTree as ET

# Directory paths (set these based on your environment)
image_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/input/images/images/welding_line"  # Directory where images (JPGs) are stored
xml_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/input/label/label"      # Directory where XML files are stored
output_directory = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/welding_line"  # Directory to save visualized images
output_zip = "/content/drive/MyDrive/Colab Notebooks/ai프로젝트/welding_line.zip"     # Output zip file path

# Ensure output directory exists
os.makedirs(output_directory, exist_ok=True)

# Function to process each pair of XML and JPG
def visualize_and_save(xml_file, image_file, output_image_path):
    # Load image
    image = Image.open(image_file)

    # Ensure the image is in RGB mode
    if image.mode != "RGB":
        image = image.convert("RGB")

    # Parse the XML annotation file
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Extract bounding box coordinates from the XML and draw them on the image
    draw = ImageDraw.Draw(image)
    for obj in root.findall('object'):
        bndbox = obj.find('bndbox')
        xmin = int(bndbox.find('xmin').text)
        ymin = int(bndbox.find('ymin').text)
        xmax = int(bndbox.find('xmax').text)
        ymax = int(bndbox.find('ymax').text)

        # Draw bounding box
        draw.rectangle([xmin, ymin, xmax, ymax], outline="red", width=3)

    # Save the modified image to the output directory
    image.save(output_image_path)

# Create a ZIP file to store all the visualized images
with zipfile.ZipFile(output_zip, 'w') as zipf:
    # Get the base names of all XML and JPG files (without extension)
    xml_files = {os.path.splitext(f)[0] for f in os.listdir(xml_directory) if f.endswith(".xml")}
    jpg_files = {os.path.splitext(f)[0] for f in os.listdir(image_directory) if f.endswith(".jpg")}

    # Find the intersection (matching files) between XML and JPG
    matching_files = xml_files.intersection(jpg_files)

    for base_name in matching_files:
        xml_file_path = os.path.join(xml_directory, base_name + ".xml")
        image_file_path = os.path.join(image_directory, base_name + ".jpg")

        # Visualize and save the image with bounding boxes
        output_image_path = os.path.join(output_directory, base_name + "_visualized.jpg")
        visualize_and_save(xml_file_path, image_file_path, output_image_path)

        # Add the visualized image to the ZIP file
        zipf.write(output_image_path, arcname=os.path.basename(output_image_path))

# Output the path to the ZIP file
print(f"Visualized images saved and zipped at: {output_zip}")
```
